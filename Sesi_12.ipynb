```python
def help_menu():
  print("\n=== HELP DOCUMENTATION ===")
  print("1. Nama: huruf saja")
  print("2. Nim: angka")
  print("3. Nilai: 0-100")
  print("==============\n")
def input_nama():
  while True:
    nama = input("masukan nama:")
    if nama.lower() == "help":
      help_menu()
    elif nama.replace(" ", " ").isalpha():
      return nama
    else:
      print("Error: nama hanya boleh huruf")
def input_nim():
  while True:
    nim = input("masukan nim:")
    if nim.lower() == "help":
      help_menu()
    elif nim.isdigit():
      return nim
    else:
      print("Error: nim harus angka")
def input_nilai():
  while True:
    nilai = input("masukan nilai: ")
    if nilai.lower() == "help":
      help_menu()
      continue
    try:
       nilai = float(nilai)
       if 0 <= nilai <= 100:
        return nilai
       else:
        print("Error: nilai harus 0-100")
    except ValueError:
      print("Error: masukan angka")

print("=== SISTEM INPUT NILAI MAHASISWA ===\n")
print("Ketik 'help' untuk bantuan\n")

nama = input_nama()
nim = input_nim()
nilai = input_nilai()

print("\n=== HASIL ===")
print("Nama:", nama)
print("Nim:", nim)
print("Nilai:", nilai)
print("====================\n")
```
